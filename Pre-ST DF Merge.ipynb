```python
#create df out of old cvs that match vars of interest
Q322_csv = pd.read_csv('Q322Records.csv')
replace_months = {'January':'1',
                  'February':'2',
                  'March':'3',
                  'July':'7',
                  'August':'8',
                  'September':'10',
                  'October':'10',
                  'November':'11',
                  'December':'12'}
Q322_csv.Date = Q322_csv.Date.replace(replace_months, regex = True)
Q322 = pd.DataFrame({'PI':Q322_csv.PINum, 
                     'Day': Q322_csv.Date.str[-8:-6], 
                     'Month': Q322_csv.Date.str[-11:-8], 
                     'Year': 2022, 
                     'Quarter': 'Q3.2022', 
                     'Num_Procedures':Q322_csv.Num_Procedures, 
                     'Hours':Q322_csv.Hours, 
                     'Charge':Q322_csv.Charge})

Q422_csv = pd.read_csv('Q422Records.csv')
Q422_csv.Date = Q422_csv.Date.replace(replace_months, regex = True)
Q422 = pd.DataFrame({'PI':Q422_csv.PINum, 
                     'Day': Q422_csv.Date.str[-8:-6], 
                     'Month': Q422_csv.Date.str[-11:-8], 
                     'Year': 2022, 
                     'Quarter': 'Q4.2022', 
                     'Num_Procedures':Q422_csv.Num_Procedures, 
                     'Hours':Q422_csv.Hours, 
                     'Charge':Q422_csv.Charge})

Q123_csv = pd.read_csv('Q123Records.csv')
Q123_csv.Date = Q123_csv.Date.replace(replace_months, regex = True)
Q123 = pd.DataFrame({'PI':Q123_csv.PINum, 
                     'Day': Q123_csv.Date.str[-8:-6], 
                     'Month': Q123_csv.Date.str[-11:-8], 
                     'Year': 2023, 
                     'Quarter': 'Q1.2023', 
                     'Num_Procedures':Q123_csv.Num_Procedures, 
                     'Hours':Q123_csv.Hours, 
                     'Charge':Q123_csv.Charge})
```

```python
#merge all dfs and clean
records = [Q122, Q222, Q322, Q422, Q123]
OR_records = pd.concat(records)
#Check : print(OR_records)
#Check : print(OR_records.dtypes)
#Check: print(OR_records[OR_records.isnull().any(axis=1)])

#one record that is all null and needs to be deleted
OR_records = OR_records.dropna()

#some minor cleaning needed due to the slicing
OR_records = OR_records.replace(' ', '', regex = True)
OR_records = OR_records.replace(',', '', regex = True)

#combine date information into a date column
OR_records['Date'] = pd.to_datetime([f'{y}-{m}-{d}' for y, m, d in zip(OR_records.Year, OR_records.Month, OR_records.Day)])

#create OR column 
def determine_OR (row):
    if row['PI'] == 1 :
        return 'BC-158'
    elif row['PI'] == 2 :
        return 'BC-158'
    elif row['PI'] == 3 :
        return 'BC-142'
    elif row['PI'] == 4 :
        return 'BC-142'
    elif row['PI'] == 5 :
        return 'BC-142'
    elif row['PI'] == 6 :
        return 'BC-158'
    else :
        return ''
OR_records['OR'] = OR_records.apply(lambda row: determine_OR(row), axis=1)

#fix data types
OR_records[['PI','Month','Year','Quarter']] = OR_records[['PI','Month','Year','Quarter']].astype('category')
OR_records['Quarter'] = OR_records['Quarter'].cat.set_categories(
    ['Q1.2022', 'Q2.2022', 'Q3.2022', 'Q4.2022', 'Q1.2023'], ordered=True)

#sort by date
OR_records.sort_values('Date')

#save file
OR_records.to_csv('OR Records Q1.2022 to Q1.2023.csv')

#print df info
print(OR_records.describe())
print(OR_records.dtypes)
print(OR_records.info())
```
